```python
import pandas as pd
import os
os.makedirs("outputs", exist_ok=True)
```

```python
df = pd.read_csv("healthcare_dataset.csv")
df.to_csv("outputs/data_extraction.csv", index=False)
print("Data extracted successfully")
df.head()
```

```python
df = pd.read_csv("data_extraction.csv")
df["Medical_Condition_Text"] = df["Medical Condition"].astype(str)
df["Text_Length"] = df["Medical_Condition_Text"].apply(len)
df.to_csv("text_processing.csv", index=False)
print("Text processing completed")
df.head()
```

```python
df = pd.read_csv("text_processing.csv")

df = df[
    [
        "Name",
        "Age",
        "Gender",
        "Blood Type",
        "Medical Condition",
        "Doctor",
        "Hospital",
        "Text_Length"
    ]
]

df.to_csv("parameter_identification.csv", index=False)

print("Key parameters identified")
df.head()
```

```python
df = pd.read_csv("parameter_identification.csv")
df.drop_duplicates(inplace=True)
df.fillna("Unknown", inplace=True)
df.to_csv("data_validation.csv", index=False)
print("Data validation completed")
df.head()
```

```python
df = pd.read_csv("data_validation.csv")
df["Age"] = pd.to_numeric(df["Age"], errors="coerce").fillna(0).astype(int)
df["Gender"] = df["Gender"].str.capitalize()
df.to_csv("data_standardization.csv", index=False)
print("Data standardization completed")
df.head()
```
