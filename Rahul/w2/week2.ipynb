```python
import pandas as pd
df = pd.read_csv("data_standardization.csv")
def identify_pattern(condition, age):
    if condition in ["Diabetes", "Hypertension"] and age >= 45:
        return "Metabolic Syndrome Pattern"
    elif condition == "Cancer":
        return "Oncology Pattern"
    elif condition == "Asthma":
        return "Respiratory Pattern"
    elif condition == "Obesity":
        return "Lifestyle Disorder Pattern"
    else:
        return "General Health Pattern"
df["Condition_Pattern"] = df.apply(
    lambda x: identify_pattern(x["Medical Condition"], x["Age"]),
    axis=1
)
df.to_csv("pattern_recognition.csv", index=False)
print("Pattern recognition completed")
df.head()
```

```python
df = pd.read_csv("pattern_recognition.csv")

def risk_level(age, condition):
    if age > 60 and condition in ["Diabetes", "Hypertension", "Cancer"]:
        return "Very High Risk"
    elif condition == "Cancer":
        return "High Risk"
    elif condition in ["Diabetes", "Hypertension"]:
        return "Moderate Risk"
    else:
        return "Low Risk"
df["Risk_Level"] = df.apply(
    lambda x: risk_level(x["Age"], x["Medical Condition"]), axis=1
)
df.to_csv("risk_assessment.csv", index=False)
print("Risk assessment completed")
df.head()
```

```python
import pandas as pd
df = pd.read_csv("risk_assessment.csv")
def contextual_risk(age, gender, risk):
    if age >= 60 and risk in ["High Risk", "Very High Risk"]:
        return "Critical Risk (Age Factor)"
    if gender == "Male" and risk == "High Risk":
        return "High Risk (Gender Factor)"
    return risk
df["Contextual_Risk"] = df.apply(
    lambda x: contextual_risk(x["Age"], x["Gender"], x["Risk_Level"]),
    axis=1
)
df.to_csv("contextual_analysis.csv", index=False)
print("Contextual analysis completed")
df.head()
```
